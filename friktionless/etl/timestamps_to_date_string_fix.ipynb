```python
import pandas as pd
from datetime import date
from tqdm.notebook import tqdm
```

```python
all_user_df = pd.read_gbq(query='select * from users.fact_user_first_deposit')
```

```python
def replace_null_times(t):
    if t == '0001-01-01T00:00:00Z':
        return pd.NaT
    else:
        return t
```

```python
date_cols = ['as_of_date', 'first_deposit_date', 'first_withdrawal_date', 'last_deposit_date', 'last_withdrawal_date', 'churn_date']
for user_address in tqdm(all_user_df.user_address.iloc[49:2000]):
    rewrite_df = pd.read_json(f'gcs://friktion-users-prod/user-{user_address}.json', orient='records', lines=True, convert_dates=date_cols)
    rewrite_df.as_of_date = rewrite_df.as_of_date.apply(date.isoformat)
    rewrite_df.first_deposit_date = rewrite_df.first_deposit_date.apply(date.isoformat)
    rewrite_df.first_withdrawal_date = rewrite_df.first_withdrawal_date.apply(replace_null_times)
    # print(rewrite_df[date_cols])
    rewrite_df.first_withdrawal_date = rewrite_df.first_withdrawal_date.apply(date.isoformat)
    rewrite_df.last_withdrawal_date = rewrite_df.last_withdrawal_date.apply(date.isoformat)
    rewrite_df.last_deposit_date = rewrite_df.last_deposit_date.apply(date.isoformat)
    rewrite_df.churn_date = rewrite_df.churn_date.apply(date.isoformat)
    rewrite_df.to_json(f'gcs://friktion-users-prod/user-{user_address}.json', orient='records', lines=True)
    print(f'gcs://friktion-users-prod/user-{user_address}.json')
```
